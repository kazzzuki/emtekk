# Import Necessary Libraries

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import imutils
import cv2
from scipy import ndimage
import time
```

# Template Image Class

```python
class Template:
    def __init__(self, image_paths, a=1, b=1):
        # Load all images and convert them to numpy arrays
        images = [np.array(Image.open(img).convert("L"), dtype=np.float32) for img in image_paths]

        # Calculate mean image
        self.image = np.mean(images, axis=0)

        # Calculate standard deviation
        self.std = np.std(images, axis=0)

        # Set empirical values a and b
        self.a = a
        self.b = b

        # Calculate light and dark images
        self.light = np.clip(self.image + np.maximum(self.a, self.b * self.std), 0, 255)
        self.dark = np.clip(self.image - np.maximum(self.a, self.b * self.std), 0, 255)

    def get_template_image(self):
        return Image.fromarray(self.image.astype(np.uint8))

    def get_light_image(self):
        return Image.fromarray(self.light.astype(np.uint8))

    def get_dark_image(self):
        return Image.fromarray(self.dark.astype(np.uint8))
```

# Image Registration and Alignment

```python
# Function to create a binary mask for defects
def get_marked_image(image_src, kernel_size, rotation=0, ideal=True):
    sample = np.asarray(image_src)
    fixed = sample if ideal else ndimage.rotate(sample, rotation)

    sample_image_gray = np.array(Image.fromarray(fixed).convert("L"), dtype=np.float32)
    sample_image_blur = cv2.GaussianBlur(sample_image_gray, (kernel_size, kernel_size), 0)

    binary_image = Image.fromarray(sample_image_blur.astype(np.uint8))

    return binary_image
```

# Demo

```python
class EdgeBasedSystem:
    def measure_latency(self, func, *args):
        start_time = time.time()
        result = func(*args)
        end_time = time.time()
        latency = (end_time - start_time) * 1000  # ms
        print(f"Edge Latency: {latency:.3f} ms")
        return result

    def measure_execution_time(self, func, *args):
        start_time = time.time()
        result = func(*args)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"Edge Execution Time: {execution_time:.3f} seconds")
        return result

    def simulate_downtime(self, down_duration):
        print("Simulating Edge System Downtime...")
        time.sleep(down_duration)  # Simulate downtime
        return down_duration

class CloudBasedSystem:
    def measure_latency(self, func, *args):
        start_time = time.time()
        result = func(*args)  # Upload to cloud, process, and return result
        end_time = time.time()
        latency = (end_time - start_time) * 1000  # ms
        print(f"Cloud Latency: {latency:.3f} ms")
        return result

    def measure_execution_time(self, func, *args):
        start_time = time.time()
        result = func(*args)  # Cloud processing
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"Cloud Execution Time: {execution_time:.3f} seconds")
        return result

    def simulate_downtime(self, down_duration):
        print("Simulating Cloud Downtime...")
        time.sleep(down_duration)  # Simulate cloud downtime (e.g., server issue)
        return down_duration

class HybridSystem(EdgeBasedSystem, CloudBasedSystem):
    def measure_latency(self, func, *args):
        # Combined latency calculation for edge and cloud processing
        edge_start_time = time.time()
        result_edge = func(*args)  # Edge processing
        edge_end_time = time.time()
        edge_latency = (edge_end_time - edge_start_time) * 1000  # ms

        cloud_start_time = time.time()
        result_cloud = func(*args)  # Cloud processing
        cloud_end_time = time.time()
        cloud_latency = (cloud_end_time - cloud_start_time) * 1000  # ms

        total_latency = edge_latency + cloud_latency
        print(f"Hybrid Latency: {total_latency:.3f} ms")
        return result_cloud

    def measure_execution_time(self, func, *args):
        # Combined execution time for edge and cloud processing
        edge_start_time = time.time()
        result_edge = func(*args)  # Edge processing
        edge_end_time = time.time()
        edge_execution_time = edge_end_time - edge_start_time

        cloud_start_time = time.time()
        result_cloud = func(*args)  # Cloud processing
        cloud_end_time = time.time()
        cloud_execution_time = cloud_end_time - cloud_start_time

        total_execution_time = edge_execution_time + cloud_execution_time
        print(f"Hybrid Execution Time: {total_execution_time:.3f} seconds")
        return result_cloud

    def simulate_downtime(self, down_duration):
        # Simulate downtime for both edge and cloud
        print("Simulating Hybrid System Downtime...")

        # Explicitly call simulate_downtime from both base classes
        print("Simulating Edge System Downtime...")
        edge_downtime = EdgeBasedSystem.simulate_downtime(self, down_duration)

        print("Simulating Cloud System Downtime...")
        cloud_downtime = CloudBasedSystem.simulate_downtime(self, down_duration)

        return edge_downtime + cloud_downtime


class HybridSystem(EdgeBasedSystem, CloudBasedSystem):
    def measure_latency(self, func, *args):
        # Combined latency calculation for edge and cloud processing
        edge_start_time = time.time()
        result_edge = func(*args)  # Edge processing
        edge_end_time = time.time()
        edge_latency = (edge_end_time - edge_start_time) * 1000  # ms

        cloud_start_time = time.time()
        result_cloud = func(*args)  # Cloud processing
        cloud_end_time = time.time()
        cloud_latency = (cloud_end_time - cloud_start_time) * 1000  # ms

        total_latency = edge_latency + cloud_latency
        print(f"Hybrid Latency: {total_latency:.3f} ms")
        return result_cloud

    def measure_execution_time(self, func, *args):
        # Combined execution time for edge and cloud processing
        edge_start_time = time.time()
        result_edge = func(*args)  # Edge processing
        edge_end_time = time.time()
        edge_execution_time = edge_end_time - edge_start_time

        cloud_start_time = time.time()
        result_cloud = func(*args)  # Cloud processing
        cloud_end_time = time.time()
        cloud_execution_time = cloud_end_time - cloud_start_time

        total_execution_time = edge_execution_time + cloud_execution_time
        print(f"Hybrid Execution Time: {total_execution_time:.3f} seconds")
        return result_cloud

    def simulate_downtime(self, down_duration):
        # Simulate downtime for both edge and cloud
        print("Simulating Hybrid System Downtime...")

        # Explicitly call simulate_downtime from both base classes
        print("Simulating Edge System Downtime...")
        edge_downtime = EdgeBasedSystem.simulate_downtime(self, down_duration)

        print("Simulating Cloud System Downtime...")
        cloud_downtime = CloudBasedSystem.simulate_downtime(self, down_duration)

        return edge_downtime + cloud_downtime
```
