### **Part 3: Calculating Correlation with Python**

**Step 1: Calculate correlation against brainFrame**

The pandas corr() method provides an easy way to calculate correlation against a dataframe. By simply
calling the method against a dataframe, one can get the correlation between all variables at the same
time.

```python
# Code cell 8
brainFrame.corr(method='pearson')
```

**Notice at the left-to-right diagonal in the correlation table generated above. Why is the diagonal filled
with 1s? Is that a coincidence? Explain.**
- No. It is not coincidence. The 1s represents the correlation of each variable with itself.


**Still looking at the correlation table above, notice that the values are mirrored; values below the 1
diagonal have a mirrored counterpart above the 1 diagonal. Is that a coincidence? Explain.**
- No. It is not coincidence either. This is because the correlation of variables A and B are the same as the correlation of variables B and A.


Using the same corr() method, it is easy to calculate the correlation of the variables contained in the
female-only dataframe:

```python
# Code cell 9
womenDF.corr(method='pearson')
```

And the same can be done for the male-only dataframe:

# **Supplementary Activity**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dengueFile = 'dengueCase.csv'
dengueFrame = pd.read_csv(dengueFile)

# VARIABLE FOR CASES
case50 = dengueFrame[dengueFrame['cases'] > 50]

# VARIABLE FOR LOCATION
locS = case50.groupby('loc')['cases'].sum()

# TABLE: LOCATION WITH > 50 DENGUE CASES + DEATHS
loc50 = case50.groupby('loc').filter(lambda x: x['cases'].sum() > 50)
displayLoc = loc50.groupby('loc').agg({'cases': 'sum', 'deaths': 'sum'}).reset_index()
print("LOCATION WITH MORE THAN 50 CASES OF DENGUE:\n")
print(displayLoc)

# TABLE: STATISTICS OF DENGUE CASES + DEATHS
print("\nSummary Statistics of Dengue Cases and Deaths (Cases > 50):\n")
print(case50.describe())

# VISUALIZATION: BAR CHART
plt.figure(figsize=(10, 6))
sns.histplot(case50['cases'], bins=10, color='blue', edgecolor='black')
plt.xlabel('Number of Dengue Cases (Cases > 50)')
plt.ylabel('Frequency')
plt.title('\nDistribution of Dengue Cases (Cases > 50)')
plt.show()

# VISUALIZATION: BAR CHART
plt.figure(figsize=(10, 6))
sns.histplot(case50['deaths'], bins=5, color='blue', edgecolor='black')
plt.xlabel('Number of Dengue Deaths (Cases > 50)')
plt.ylabel('Frequency')
plt.title('\nDistribution of Dengue Deaths (Cases > 50)')
plt.show()

# VISUALIZATION: PIE CHART
plt.figure(figsize=(8, 8))
plt.pie(locS, labels=locS.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Dengue Cases by Location (Cases > 50)\n')
plt.axis('equal')
plt.show()

# VISUALIZATION: DOT CHART
plt.figure(figsize=(10, 6))
plt.plot(locS.values, locS.index, 'o', color='blue', markersize=8)
plt.xlabel('Dengue Cases')
plt.ylabel('Location')
plt.title('\nProportion of Dengue Cases by Location (Cases > 50)')
plt.grid(axis='x')
plt.show()
```
